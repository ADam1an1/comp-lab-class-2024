```python
with open('Data/1hz3_T310.run.25000000.energy.xvg') as f:
    data = f.read()
    data = data.split('\n')

print(data)
```

```python
np_data = data[29:-1]
print(np_data[0])
print(np_data[-1])
labels_data = data[23:29]
units_data = data[15].split('"')[1].split(',')[0:1]*3 + data[15].split('"')[1].split(',')
print(labels_data)
print(units_data)
```

```python
np_data_split = [line.split(' ') for line in np_data]
np_data_clean = np.zeros(shape=(501, 7))
for i in range(0, len(np_data_split)):
    np_data_clean[i] += [float(j) for j in np_data_split[i] if j != '']
```

```python
import matplotlib.pyplot as plt

# Part 4a
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
for i in range(1, len(np_data_clean[0])):
    color = colors[i - 1]
    plt.plot(np_data_clean[:, 0]/1000, np_data_clean[:, i], color)
    plt.xlabel('Time (ns)')
    plt.ylabel(labels_data[i - 1].split("\"")[1] + ' ' + units_data[i - 1])
    plt.title(labels_data[i - 1].split("\"")[1] + ' against Time')
    plt.savefig('./Figures/' + 'plot_' +  labels_data[i - 1].split("\"")[1].lower() + '_time.png')
    plt.show()
```

```python
# Part 4b

fig, ax1 = plt.subplots()
ax1.set_xlabel('Time (ns)')

color = colors[-1]
index = 4
plt.ylabel(labels_data[index - 1].split("\"")[1] + ' ' + units_data[index - 1])
ax1.plot(np_data_clean[:, 0]/1000, np_data_clean[:, index], color=color)

ax2 = ax1.twinx()

color = colors[0]
index = 2
plt.ylabel(labels_data[index - 1].split("\"")[1] + ' ' + units_data[index - 1])
ax2.plot(np_data_clean[:, 0]/1000, np_data_clean[:, index], color=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
fig.legend([labels_data[1].split("\"")[1], labels_data[3].split("\"")[1]], loc = 'upper center')

plt.savefig('./Figures/Part 4/' + 'plot_kinetic_temp_time.png')
plt.show()
```

```python
# Part 5a
import mdtraj
traj = mdtraj.load('Data/1UBQ_processed.pdb')
print(traj)
```

```python
# Part 5b
hbond_bh = mdtraj.baker_hubbard(traj, freq=0.1, exclude_water=True, periodic=True, sidechain_only=False)
hbond_wn = mdtraj.wernet_nilsson(traj, exclude_water=True, periodic=True, sidechain_only=False)
print('hbonds via baker hubbard: ', len(hbond_bh))
print('hbonds via wernet nilsson: ', np.shape(hbond_wn)[1])
```

```python
# Part 5c
distinction = mdtraj.compute_dssp(traj, simplified=False)
count = np.count_nonzero(distinction == 'H')
print(count)
```

```python
# Part 6a
traj = mdtraj.load('Data/1hz3_T310.stepid25000000.every100ps.nowater.xtc', top='Data/1hz3_T310.start.nowater.gro')
print(traj)
```

```python
# Part 6b
end_end = mdtraj.compute_distances(traj, [[0, 407]], periodic=True, opt=True)
gyr_rad = mdtraj.compute_rg(traj)
```

```python
# Part 6c
fig, ax1 = plt.subplots()
ax1.set_xlabel('Time (ps)')

index = 4
plt.ylabel('End-End Distance (nm)')
ax1.plot(traj.time / 10, end_end, color = colors[0])

ax2 = ax1.twinx()

index = 2
plt.ylabel('Gyration Radius (nm)')
ax2.plot(traj.time / 10, gyr_rad, color = colors[1])

fig.tight_layout()  # otherwise the right y-label is slightly clipped
fig.legend([labels_data[1].split("\"")[1], labels_data[3].split("\"")[1]], loc = 'upper center')
plt.savefig('./Figures/plot_ubiquitin_end_gyr_time.png')
plt.show()
```

```python
# Part 6d
plt.hist(end_end, stacked=True, density=True)
plt.xlabel('End-End Distance (nm)')
plt.ylabel('Density')
plt.title('Normalized Histogram End-End Distance')
plt.savefig('./Figures/plot_ubiquitin_end_hist.png')
plt.show()

plt.hist(gyr_rad, stacked=True, density=True, color = colors[1])
plt.xlabel('Gyration Radius(nm)')
plt.ylabel('Density')
plt.title('Normalized Histogram Gyration Radius')
plt.savefig('./Figures/plot_ubiquitin_gyr_hist.png')
plt.show()
# plt.hist(end_end)
```

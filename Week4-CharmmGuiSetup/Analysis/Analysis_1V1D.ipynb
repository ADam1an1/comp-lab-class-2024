```python
import os, sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
# RMSD of MD

x,y = np.loadtxt("1V1D_rmsd.xvg", comments=["@", "#"], unpack=True)
plt.plot(x, y, label="Ref: Equilibrated")

x,y = np.loadtxt("1V1D_rmsd_xtal.xvg", comments=["@", "#"], unpack=True)
plt.plot(x, y, label="Ref: PDB")

plt.xlabel('Time (ns)')
plt.ylabel('RMSD (nm)')
plt.title('RMSD of Structure')
plt.legend()
plt.savefig("../Figures/1V1D_rmsd.png")
```

```python
# MD Gyration

x_gyr,y_sum,y_x,y_y,y_z = np.loadtxt("1V1D_gyrate.xvg", comments=["@", "#"], unpack=True)
plt.plot(x_gyr, y_sum)

plt.xlabel('Time (ns)')
plt.ylabel('R_g (nm)')
plt.title('Radius of Gyration')
plt.savefig("../Figures/1V1D_gyration.png")
```

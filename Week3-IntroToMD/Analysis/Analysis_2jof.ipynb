```python
# Energy Minimization Potential Plot
import numpy as np
import matplotlib.pyplot as plt

x,y = np.loadtxt("2jof_em_potential.xvg", comments=["@", "#"], unpack=True)
plt.xlabel('Energy Minimization Step')
plt.ylabel('Potential Energy (kJ mol^-1)')
plt.title('Steepest Descent Energy Minimization')
plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
plt.plot(x, y)
plt.savefig("../Figures/2jof_em_potential.png")
```

```python
# Temperature plot for NVT equilibration

x,y = np.loadtxt("2jof_100ps_nvt_temp.xvg", comments=["@", "#"], unpack=True)
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.title('NVT Equilibration Temperature')
plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
plt.plot(x, y)
plt.savefig("../Figures/2jof_nvt_temp.png")
```

```python
# Pressure plot for NPT equilibration

x,y = np.loadtxt("2jof_100ps_nvtnpt_press.xvg", comments=["@", "#"], unpack=True)
plt.xlabel('Time (ps)')
plt.ylabel('Pressure (bar)')
plt.title('NPT Equilibration Pressure')
window = 10
average_y = []
for ind in range(len(y) - window +1):
    average_y.append(np.mean(y[ind:ind+window]))
for ind in range(window - 1):
    average_y.insert(0, np.nan)
plt.plot(x, y)
plt.plot(x, average_y, label='10-ps Running Avg.')
plt.legend()
plt.savefig("../Figures/2jof_nvtnpt_press.png")
```

```python
# Density plot for NPT equilibration

x,y = np.loadtxt("2jof_100ps_nvtnpt_dens.xvg", comments=["@", "#"], unpack=True)
plt.xlabel('Time (ps)')
plt.ylabel('Density (kg m^-3))')
plt.title('NPT Equilibration Density')
window = 10
average_y = []
for ind in range(len(y) - window +1):
    average_y.append(np.mean(y[ind:ind+window]))
for ind in range(window - 1):
    average_y.insert(0, np.nan)
plt.plot(x, y)
plt.plot(x, average_y, label='10-ps Running Avg.')
plt.legend()
plt.savefig("../Figures/2jof_nvtnpt_dens.png")
```
